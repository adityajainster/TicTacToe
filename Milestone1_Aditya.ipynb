```python
def user_choice_player1():
    
    #Initial
    choice = 'WRONG'
    acceptable_range = range(1,10)
    within_range = False
    
    #THREE CONDITIONS TO CHECK
    #DIGIT OR WITHIN RANGE==FALSE
    #CHOICE POSITION IS IN AN EMPTY CELL
    while choice.isdigit()==False or within_range==False or board_list[int(choice)]!=' ':
    
        choice = input("Please enter a position of 'X' for Player 1:")
        
        #DIGIT CHECK
        if choice.isdigit()==False:
            print("Sorry that is not a digit!")
        
        #RANGE CHECK
        if choice.isdigit()==True:
            if int(choice) in acceptable_range:
                within_range=True
            else:
                print("Sorry, you are out of range- 0 to 10")
                within_range=False
                
        #EMPTY CELL CHECK
        if board_list[int(choice)]!=' ':
            print("Cell is already filled - Enter a different choice")
        else:
            pass
    return int(choice)
```

```python
def x_choice(board_list,position):
      
    board_list[position] = 'X'
    
    return board_list
```

```python
def user_choice_player2():
    
    #Initial
    choice = 'WRONG'
    acceptable_range = range(1,10)
    within_range = False
    
    #THREE CONDITIONS TO CHECK
    #DIGIT OR WITHIN RANGE==FALSE
    #CHOICE POSITION IS IN AN EMPTY CELL
    while choice.isdigit()==False or within_range==False or board_list[int(choice)]!=' ':
    
        choice = input("Please enter a position of 'O' for Player 2:")
        
        #DIGIT CHECK
        if choice.isdigit()==False:
            print("Sorry that is not a digit!")
        
        #RANGE CHECK
        if choice.isdigit()==True:
            if int(choice) in acceptable_range:
                within_range=True
            else:
                print("Sorry, you are out of range- 0 to 10")
                within_range=False
                
        #EMPTY CELL CHECK
        if board_list[int(choice)]!=' ':
            print("Cell is already filled - Enter a different choice")
        else:
            pass
    return int(choice)
```

```python
def o_choice(board_list,position):
      
    board_list[position] = 'O'
    
    return board_list
```

```python
def gameon_choice():
    choice = 'wrong'
    while choice not in ['Y','N']:
        choice  =input("Keep Playing? (Y or N)")
        
        if choice not in ['Y','N']:
            print("Sorry, invalid choice - Please choose Y or N")
            
        if choice=="Y":
            return True
        else:
            return False
    
    return int(choice)
```

```python
def check_win_combinations(win_combinations):
    for combination in win_combinations:
        if combination == ['X','X','X']:
            player1_wins = True
        elif combination == ['O','O','O']:
            player2_wins = True
        else:
            pass
```
